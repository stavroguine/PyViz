```python
import requests 
import pandas as pd
import json
from pandas.io.json import json_normalize

baseUrl = "https://opendata.lillemetropole.fr/api/records/1.0/search/?dataset="
paramUrl = "&rows=120"

services = { "bureaux-de-poste",
            #"localisation-hopitaux",
            #"magasins-et-boutiques",
            #"cinemas-theatres-mel",
            #"adresse-et-geolocalisation-des-etablissements-denseignement-du-premier-et-second",
}

for service in services:
 urlApi = baseUrl + service + paramUrl
 urlCall = json.loads(requests.get(urlApi).text)
 dataFrame  = pd.DataFrame.from_dict(pd.io.json.json_normalize(urlCall['records']), orient='columns')
 print (dataFrame)
```
